```python
%%writefile code.js
// Part A
const data = [5, 12, 3, 20, 11, 9, 30];
const output = []
for (let i = 0; i < data.length; i++){
    if (data[i] > 10) {
        output.push(data[i] * 2);
    }
}
console.log(output);

const output2 = data.filter(num => num > 10).map(num => num * 2);
console.log(output2);

// Part B

const numbers = [1, 2, 3, 4, 5, 6];

const sumOfDoubledEvens = numbers.filter(num => num % 2 === 0).map(num => num * 2).reduce((acc, num) => acc + num, 0);

console.log(sumOfDoubledEvens);

// Part C
const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];

console.log(users.some(users => users.age > 18));
console.log(users.find(users => users.id === 3));
users.forEach(users => console.log(users.name));


// Part D

function repeat(n, action) {
  // TODO: call action(i) for i = 0..n-1
  for (let i = 0; i < n; i++) {
    action(i);
  }
}

repeat(3, console.log); // 0, 1, 2
repeat(4, i => console.log("Step " + i));
```
