```python
%%writefile code.js
function addTax(price, rate){
  return price + (price * rate);
}
console.log("Add Tax:");
console.log(addTax(10, 0.5));
console.log(addTax(100, 0.1));

function createBurner(secret, limit){
  return () => {
    if(limit > 0){
      limit--;
      return secret;
    }else{
      return "Limit Reached";
    }
  }
}

console.log("Burner:")
const burn = createBurner("rock", 3);
console.log(burn());
console.log(burn());
console.log(burn());
console.log(burn());
```
